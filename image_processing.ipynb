How to resize the images using pillow library ?

```python
from PIL import Image
```

```python
img=Image.open('/content/Dog')
img_resized=img.resize((200,200))
```

```python
img_resized.save('dog_img_resized.jpg')
```

```python
img_new= mpimg.imread('/content/dog_img_resized.jpg')
img_new_plot=plt.imshow(img_new)
plt.show()
```

Coverting rbg images into gray scale image using open cv

```python
#import opencv library

import cv2
```

```python
img =cv2.imread('/content/dog_img_resized.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
gray_img= cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```
